# Fundamentos de bases de datos y para ciencia de datos
Una base de datos es una solución tecnológica que nos permite registrar un conjunto de datos estructurados, almacenarlos en la memoria de una computadora o en la nube, accederlos de varias maneras y recuperarlos de forma segura, precisa, eficiente y rápida. Adicional de tener otras ventajas como la facilidad de intercambiar información y sincronizar análisis. Existen varios  tipos de bases de datos:
-	Relacionaleso SQL: los datos se organizan y almacenan en tablas que se pueden vincular entre sí utilizando alguna relación. La forma de comunicarse e interactuar con bases de datos relacionales es mediante el uso del lenguaje SQL.

-	No relacionales o NoSQL: Estas bases de datos son aquellas que conectan la información almacenada en ellas por categorías en lugar de relaciones. Como su nombre lo indica estas bases de datos no usan SQL para comunicarse, la forma más común son los pares clave-valor.

-	Distribuidas: Este tipo de bases de datos hacen referencia a la que puede ser una bases de datos que están en múltiples ubicaciones/sitios/computadoras

-	Cadena de bloques o BlockChain: Este tipo de base de datos hace relación a la autoridad y propiedad de la base de datos no es única. 


# Fundamentos de almacenes de datos (Data Warehouse) para ciencia de datos.
Un almacén de datos es un tipo de sistema de gestión de datos que está diseñado para habilitar y respaldar actividades de inteligencia empresarial. Básicamente, están destinados únicamente a realizar consultas y análisis. Los datos dentro de un almacén de datos generalmente se derivan de una amplia gama de fuentes, como archivos de registro de aplicaciones y aplicaciones de transacciones que pueden estar en varias bases de datos. Dentro de cada base de datos, los datos se organizan en tablas y en columnas. Dentro de cada columna, se puede definir una descripción de los datos, como un número entero, un campo de datos o una cadena. Las tablas se pueden organizar dentro de esquemas, que se pueden concebir como carpetas. Cuando los datos se incorporan, se almacenan en varias tablas descritas por el esquema. Las herramientas de consulta usan el esquema para determinar a qué tablas de datos obtendrán acceso y cuáles de ellas analizarán.

*Beneficios*

- Toma de decisiones fundamentadas
- Datos consolidados provenientes de muchos orígenes
- Análisis de datos históricos
- Calidad, coherencia y precisión de los datos
- Separación del procesamiento de análisis de las bases de datos transaccionales, lo que mejora el rendimiento de ambos sistemas
